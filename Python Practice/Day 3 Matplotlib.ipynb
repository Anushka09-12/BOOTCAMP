```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,10])

y = np.array([1,50])

plt.plot(x,y)
```

```python
y_points = np.array([1,10,2,8,5])
```

```python
plt.plot(y_points)
```

```python
plt.plot(y_points,marker="o",linestyle="dotted")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sample")
```

```python
months = np.array(['January', 'February', 'March', 'April', 'May'])
sales = np.array([2500, 3200, 2800, 4100, 3700])
```

```python
plt.bar(months,sales,color="purple")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Data")
```
